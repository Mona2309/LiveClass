```python
import pandas as pd
import mysql.connector as conn
import pyodbc



# Import CSV
data = pd.read_csv (r'C:\Users\monalisha.dhal\cardata.csv')
print(data.head())
df = pd.DataFrame(data)
print(df.head())

print(data)
# Connect to SQL Server
mydb = conn.connect(database = "cardataset" , host = "localhost", user = "root", passwd = "Welcome@123")
cursor = mydb.cursor()
#cursor.execute("use cardataset")

cursor.execute("DROP TABLE IF EXISTS Car1")

# Create Table
cursor.execute("""CREATE TABLE Car1 (
                  Buying  VARCHAR(20) ,
                   Maint VARCHAR(50),
                   doors VARCHAR(50),
				   persons VARCHAR(50),
                   lug_boot VARCHAR(5),
                   safety VARCHAR(5),
				   classdet VARCHAR(50))
				   """)
                   


# Insert DataFrame to Table

df.columns = df.columns.str.strip() 
print(df.columns)

for i in range(len(df)):
    print(df.loc[i])
    cursor.execute('''
                INSERT INTO cardataset.Car1(Buying, Maint, doors,persons,lug_boot,safety,classdet)
                VALUES (%s, %s, %s, %s, %s, %s,%s)
                ''',(
                df.loc[i, 'Buying'], 
                df.loc[i, 'Maint'],
                df.loc[i, 'doors'],
                df.loc[i, 'persons'],
                df.loc[i, 'lug_boot'],
                df.loc[i, 'safety'],
                df.loc[i, 'classdet']
                ))
mydb.commit()
```
